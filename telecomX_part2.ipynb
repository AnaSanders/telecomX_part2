# Interpretação e Conclusões



### **1. Análise Exploratória e Identificação de Fatores de Evasão**

#### **Desequilíbrio de Classes**
A primeira análise revelou que a base de dados é desbalanceada:

* **Clientes que Evadiram (Churn):** 26.58%
* **Clientes que Permaneceram:** 73.42%

#### **Análise de Correlação e Visualizações**

**Principais Fatores que AUMENTAM a Evasão:**

* **Tipo de Contrato:** A análise de correlação e o histograma mostram de forma conclusiva que o contrato do tipo "Month-to-month" (Mensal) é o principal fator de evasão. Clientes sem um vínculo de longo prazo têm muito mais propensão a cancelar.
* **Serviço de Internet de Fibra Ótica:** Curiosamente, ter o serviço de fibra ótica apresentou uma correlação positiva com o churn. Isso pode indicar problemas com o preço, a estabilidade do serviço ou expectativas não atendidas pela tecnologia.
* **Método de Pagamento com Cheque Eletrônico:** Clientes que utilizam cheque eletrônico como forma de pagamento também mostraram uma maior tendência à evasão.
* **Cobrança Mensal:** Contas mensais mais altas estão positivamente correlacionadas com o churn.

**Principais Fatores que DIMINUEM a Evasão (Retenção):**

* **Tempo de Serviço:** A correlação negativa mais forte foi com o tempo de serviço em meses. Clientes mais antigos e leais raramente cancelam o serviço, como ilustrado no box plot.
* **Contratos de Longo Prazo:** Contratos de um ou dois anos são fortes indicadores de retenção.
* **Suporte Técnico:** Clientes que possuem o serviço de suporte técnico tendem a permanecer mais na base.

### **2. Modelagem Preditiva e Avaliação**

* **Dummy Classifier:** Serviu como nosso modelo de base (baseline). Ele prevê sempre a classe majoritária (não evasão) e alcançou uma acurácia de 73.4%, que é o nosso ponto de partida.
* **Árvore de Decisão:** Um modelo simples e de fácil interpretação. Com uma profundidade máxima de 3 para evitar overfitting, este modelo foi superior ao baseline.
* **K-Nearest Neighbors (KNN):** Um modelo baseado em proximidade. Os dados foram normalizados antes do treinamento, o que é uma boa prática para algoritmos sensíveis à escala das variáveis.

A tabela abaixo resume o desempenho dos modelos no conjunto de teste:

| Métrica | Dummy Classifier | Árvore de Decisão | KNN |
| :--- | :---: | :---: | :---: |
| **Acurácia** | 73.4% | 78.9% | 75.7% |
| **Precisão** | 0.0% | 69.2% | 54.5% |
| **Recall** | 0.0% | 36.9% | 51.3% |
| **F1-Score** | 0.0% | 48.1% | 52.9% |

**Análise de Desempenho:**

* A **Árvore de Decisão** obteve a maior **acurácia** e **precisão**. Isso significa que, quando ela prevê que um cliente vai evadir, ela tem uma alta chance de estar correta.
* O modelo **KNN** apresentou o maior **Recall** e **F1-Score**. O Recall mais alto indica que ele foi o melhor modelo em identificar os clientes que de fato iriam cancelar.
* O **F1-Score**, que é uma média harmônica entre precisão e recall, confirma que o **KNN** oferece o melhor equilíbrio geral para este problema.
* As matrizes de confusão geradas visualizam esse comportamento, mostrando que o KNN identificou mais "Verdadeiros Positivos" (evasões corretas) que a Árvore de Decisão.

### **3. Conclusões e Estratégias de Retenção**
Com base na análise exploratória e no desempenho dos modelos, concluímos que os principais fatores que levam um cliente da TelecomX à evasão são:

1.  **Contratos Mensais (Month-to-month):** Clientes sem um compromisso de longo prazo são os mais voláteis.
2.  **Tempo de Serviço Curto:** A maior parte da evasão ocorre nos primeiros meses de contrato.
3.  **Cobranças Mensais Elevadas:** Clientes com faturas mais altas tendem a cancelar mais.
4.  **Serviço de Fibra Ótica e Pagamento via Cheque Eletrônico:** Ambos são fatores secundários, mas relevantes, que merecem investigação.

**estratégias de retenção**:

* **Campanhas de Migração de Contrato:** Criar ofertas atrativas (descontos, upgrade de serviço) para incentivar clientes do plano mensal a migrarem para contratos de 1 ou 2 anos. O modelo preditivo pode ser usado para direcionar essas ofertas aos clientes com maior probabilidade de evasão.
* **Programa de Onboarding para Novos Clientes:** Desenvolver um programa de boas-vindas e acompanhamento nos primeiros 3 a 6 meses de contrato, oferecendo suporte proativo, tutoriais de uso dos serviços e verificando a satisfação para fortalecer o relacionamento inicial.
* **Análise de Preços e Serviços de Fibra Ótica:** Realizar uma investigação específica sobre a evasão de clientes de fibra. O problema pode ser o preço, a qualidade da instalação ou a estabilidade da conexão. Uma pesquisa de satisfação focada neste público pode revelar as causas.
* **Incentivo a Métodos de Pagamento Automático:** Oferecer pequenos descontos ou benefícios para clientes que migrarem do cheque eletrônico para o pagamento automático (cartão de crédito ou débito em conta), que está associado a uma maior retenção.
